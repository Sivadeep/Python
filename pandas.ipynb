```python
import pandas as pd
```

```python
import random
l=[]

ages=[23,45,43,46]
salaries=[700000,600000,500000,1000000]
names=['ashok','anil','pavan','jagadish','vasu']
for i in range(20):
    l.append({'id':i,
             'name':random.choice(names),
             'sal':random.choice(salaries),
             'age':random.choice(ages)})
    
```

```python
l
```

```python
df = pd.DataFrame(l)
```
